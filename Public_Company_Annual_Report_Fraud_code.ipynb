## One class SVM for bank
From above bar chart, it is obvious that bank Industry has no fraud records. In addition, the missing values of bank is much higher than that of other industries. In this case, we need to build a seperate model for bank, which is One class SVM

```python
bank = data[data.Industry=='bank']
bank.head()
```

```python
# check missing value of bank industry
def missing_values_table(df):
    mis_val = df.isnull().sum() 
    mis_val_percent = 100 * df.isnull().sum() / len(df) 
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis = 1) 
    mis_val_table_ren_columns = mis_val_table.rename(columns = {0:'Missing Values',
                                                               1:'% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values',ascending=False).round(1)    
    return mis_val_table_ren_columns
missing_values_table(bank)
```

```python
bank_X = bank.iloc[:,5:]
bank_y=bank.Fraud
#delete the columns that has missing value percentage is over 70%, as it is meaningless for prediction
bank_del = missing_values_table(bank_X)[missing_values_table(bank_X).iloc[:,1] > 70.0].index 
bank_X.drop(bank_del,axis=1,inplace=True)
```

```python
#fill the rest features with mean value
for column in bank_X.columns:
    bank_X[str(column)]=bank_X[str(column)].fillna(bank_X[str(column)].mean())
```

```python
#train one class svm model for bank industry
from sklearn.svm import OneClassSVM
clf = OneClassSVM(kernel='rbf',gamma='auto')
clf.fit(bank_X)
```

## Baseline:Dummy,Logistic

```python
#Dummy classifier baseline 
from imblearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,plot_confusion_matrix,confusion_matrix,roc_curve,auc,accuracy_score,recall_score
from sklearn.metrics import f1_score as prc
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from mlxtend.plotting import plot_decision_regions, plot_confusion_matrix
from matplotlib import pyplot as plt

dummy = DummyClassifier()
dummy.fit(X_train,y_train)
y_pred_valid=dummy.predict(X_valid)
print(prc(y_valid,y_pred_valid))
print(classification_report(y_valid,y_pred_valid))
```

```python
# logistic baseline
lr = LogisticRegression()
 
lr.fit(X_train, y_train)
 
y_pred_valid = lr.predict(X_valid)
 
print(prc(y_valid,y_pred_valid))
print(classification_report(y_valid,y_pred_valid))
plot_confusion_matrix(confusion_matrix(y_valid,y_pred_valid))
```

## SMOTE

```python
from imblearn.over_sampling import SMOTE



pipe = make_pipeline(
    SMOTE(random_state=42),
    LogisticRegression()
)
 
pipe.fit(X_train,y_train)
y_valid_pred = pipe.predict(X_valid)
 
# Evaluate the model
print(prc(y_valid,y_valid_pred))
print(classification_report(y_valid,y_valid_pred))
plot_confusion_matrix(confusion_matrix(y_valid,y_pred_valid))
```

## RandomOverSampler

```python
from imblearn.over_sampling import RandomOverSampler

pipe = make_pipeline(
    RandomOverSampler(random_state=42),
    LogisticRegression()
)
 
pipe.fit(X_train,y_train)
y_valid_pred = pipe.predict(X_valid)
 
# Evaluate the model
print(prc(y_valid,y_valid_pred))
print(classification_report(y_valid,y_valid_pred))
plot_confusion_matrix(confusion_matrix(y_valid,y_pred_valid))
```

## SMOTETomek

```python
from imblearn.combine import SMOTETomek

pipe = make_pipeline(
    SMOTETomek(random_state=42),
    LogisticRegression()
)
 
pipe.fit(X_train,y_train)
y_valid_pred = pipe.predict(X_valid)
 
# Evaluate the model
print(prc(y_valid,y_valid_pred))
print(classification_report(y_valid,y_valid_pred))
plot_confusion_matrix(confusion_matrix(y_valid,y_pred_valid))
```

## Winner model: SMOTETomek

```python
# get the recall score under different ratio
from imblearn.combine import SMOTETomek
score = []
percent = []
for i in np.linspace(0.7,1,7):
    pipe = make_pipeline(
        SMOTETomek(i,random_state=42),
        LogisticRegression())
    pipe.fit(X_train,y_train)
    y_valid_pred = pipe.predict(X_valid)
    score.append(recall_score(y_valid_pred,y_valid))
    percent.append(i)
tunning = pd.DataFrame(percent,score)
tunning =tunning.reset_index()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,plot_confusion_matrix,confusion_matrix,roc_curve,auc,precision_recall_curve
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score as prc
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
weights = np.linspace(0.05, 0.95, 20)

params={
        'logistic':{
            'estimator':LogisticRegression(solver='liblinear',random_state=42),
            'param_grid':{
                'class_weight': [{0: x, 1: 1.0- x} for x in weights],
                 'C':np.linspace(0.001,100,10),
                 'penalty':['l1','l2']
            }},
        'mlp':{
            'estimator':MLPClassifier(random_state=42),
            'param_grid':{
            "hidden_layer_sizes": [(i*5,)*i for i in range(5,10)],
            'activation':['relu','elu','selu'],
            'solver':['sgd','adam'],
            'alpha':np.linspace(0.001,100,10)
            }},
        'decisiontree':{
           'estimator':DecisionTreeClassifier(random_state=42),
           'param_grid':{
            'min_samples_leaf':[2,3,5,10],
            'max_features':np.arange(5,20,1),
            'max_depth':np.arange(5,20,1)
           }},

        'randomforest':{
           'estimator':RandomForestClassifier(random_state=42),
           'param_grid':{
            'n_estimators':np.arange(5,20,1),
            'max_features':np.arange(5,20,1),
            'max_depth':np.arange(2,10,1)
           }},
       'rbfsvm':{
           'estimator':SVC(kernel='rbf',random_state=42),
           'param_grid':{
            'C':np.linspace(0.001,100,10),
            'gamma':[0.000001, 0.00001,0.00005,0.00008,0.0001]
           }},
        'linearsvc':{
            'estimator':LinearSVC(),
            'param_grid':{
            'class_weight': [{0: x, 1: 1.0- x} for x in weights],
            'C':np.linspace(0.001,100,10),
            'penalty':['l1','l2']
            }}
}
```

```python
class Training:
    
    def __init__(self):
        self.params=params
    
    def fit(self,X_scaled,y_train_os,model,plot=True):
        gs=GridSearchCV(**self.params[model],n_jobs=-1,
                        verbose=1)
        gs.fit(X_scaled,y_train_os)
        precision,recall,thresholds = precision_recall_curve(y_valid,gs.predict(X_valid_scaled))
        
        if plot:
            plt.plot(recall, precision)
            plt.xlabel('recall')
            plt.ylabel('precision')
            plt.title('Precision-recall curve :{}'.format(model))
            plot_confusion_matrix(gs, X_valid_scaled,y_valid)
            plt.rcParams['figure.figsize'] = (4.0, 4.0)
            print('{}'.format(model))
            print(classification_report(y_valid,gs.predict(X_valid_scaled)))
            print('Best F1 score:{:.2f}'.format(prc(y_valid,gs.predict(X_valid_scaled))))
        return gs
  

t=Training()
```
